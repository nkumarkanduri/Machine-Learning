```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Padma_Shri_award_recipients_(2010%E2%80%932019)'
response = requests.get(url)
soup = BeautifulSoup (response.text,'html.parser')
```

```python
table = soup.find('table',{'class':'wikitable plainrowheaders sortable'}).tbody

rows = table.find_all('tr')

columns = [v.text.replace("\n","") for v in rows[0].find_all('th')]
print(columns)
```

```python
df = pd.DataFrame(columns = columns)
```

```python
for i in range(1, len(rows)):
    tds = rows[i].find_all('td')
    if len(tds) == 3:
        values =[tds[0].text.replace("\n",''),'',tds[1].text,tds[2].text.replace("\n",'')]
    else:
        values =[td.text.replace("\n",'') for td in tds]
    df = df.append(pd.Series(values, index = columns), ignore_index = True)
```

```python
links = table.findAll('a')
Names =[]

for link in links:
    Names.append(link.get('title'))
```
